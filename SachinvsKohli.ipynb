```python
#player runs excluding sachin
ex_sac_df = sachin_df[sachin_df['Innings Player'] != 'SR Tendulkar']
RPI_ex_sac_df = ex_sac_df['Innings Runs Scored Num'].sum()/len(ex_sac_df)
```

```python
#player runs excluding kohli
ex_kohli_df = Kohli_df[Kohli_df['Innings Player'] != 'V Kohli']
RPI_ex_kohli_df = ex_kohli_df['Innings Runs Scored Num'].sum()/len(ex_kohli_df)
```

```python
#Sachin vs Rest RPI
Sachin_vs_Rest_RPI = sdf_runs_per_innings/RPI_ex_sac_df 
```

```python
#Kohli vs Rest RPI
Kohli_vs_Rest_RPI = vdf_runs_per_innings/RPI_ex_kohli_df 
```

```python
print(Sachin_vs_Rest_RPI,Kohli_vs_Rest_RPI )
```

```python
#Sachin vs Rest Strike Rate
SR_ex_sac = 100*ex_sac_df['Innings Runs Scored Num'].sum() / ex_sac_df['Innings Balls Faced'].sum()
Sachin_vs_Rest_SR = sdf_strike_rate/SR_ex_sac
```

```python
#Kohli vs Rest Strike Rate
SR_ex_kohli = 100*ex_kohli_df['Innings Runs Scored Num'].sum() / ex_kohli_df['Innings Balls Faced'].sum()
Kohli_vs_Rest_SR = vdf_strike_rate/SR_ex_kohli
```

```python
print (SR_ex_sac, SR_ex_kohli)
print (sdf_strike_rate, vdf_strike_rate)
print (Sachin_vs_Rest_SR , Kohli_vs_Rest_SR)
```

```python
# Sachin vs Rest (Matches per 100)

Sachin_mat_per_100= len(sdf)/ sum(sdf['100 s'])
Rest_mat_per_100_sac = len(ex_sac_df)/sum(ex_sac_df['100 s'])
print (Sachin_mat_per_100,Rest_mat_per_100_sac)
```

```python
# Kohli vs Rest (Matches per 100)

Kohli_mat_per_100= len(vdf)/ sum(vdf['100 s'])
Rest_mat_per_100_kohli = len(ex_kohli_df)/sum(ex_kohli_df['100 s'])
print (Kohli_mat_per_100,Rest_mat_per_100_kohli)
```
